```python
# Import packages
import os
import pandas as pd
import numpy as np
```

```python
# Import excel files

RFPFile = pd.ExcelFile("RFP2017 All.xlsx")
ResponseFile = pd.ExcelFile("AttainSolutions2016 All.xlsx")

RFP_df = RFPFile.parse()
Response_df = ResponseFile.parse()

# Replace nan values with "-"
RFP_df.fillna("-", inplace=True)
Response_df.fillna("-", inplace=True)
```

```python
# use iterrows()to iterate over each row of Response dataframe
for i, row in Response_df.iterrows():
    # compare the "requirement description" in the response sheet row to the requirement description column in RFP spreadsheet
    match = RFP_df.loc[RFP_df["Requirement Description"] == row["Requirement Description"]]
    # if the output index is not of size zero copy response sheet stuff over to RFP sheet
    if match.index.size != 0:
        # note: here match.index returns the row number within RFP_df that matches with the Response_df requirement description
        RFP_row_num = match.index[0]
        # for columns 3 to 6, copy contents of the Response_df (represented by the row variable) into the RFP_df
        for n in range(3,6):
            RFP_df.iloc[[RFP_row_num],n] = str(row[n])
        
print(RFP_df)
    
```

```python
# output the files 

writer = pd.ExcelWriter("output1.xlsx", engine='xlsxwriter')
RFP_df.to_excel(writer)
writer.save()
```
